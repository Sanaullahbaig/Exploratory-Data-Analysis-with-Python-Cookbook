Import relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load dataset

```python
penguins_data = pd.read_csv("penguins_size.csv")
```

Subset for relevant columns

```python
penguins_data = penguins_data[['species','culmen_length_mm']]
```

Inspect first 5 rows and data types of the dataset

```python
penguins_data.head()
```

```python
penguins_data.shape
```

```python
penguins_data.dtypes
```

Group our data to get count of species

```python
penguins_group = penguins_data.groupby('species').count()
penguins_group
```

Reset the index to ensure the index isn't the species column

```python
penguins_groups = penguins_group.reset_index()
penguins_groups
```

Create a pie chart in matplotlib

```python
plt.pie(penguins_groups["culmen_length_mm"], labels = penguins_groups['species'])
plt.show()
```

Provide additional chart details

```python
cols = ['g', 'b', 'r'] # colors green, blue and red

plt.pie(penguins_group["culmen_length_mm"], labels = penguins_groups['species'],colors = cols)

plt.title('Univariate Analysis of Species', fontsize=15)
plt.show()
```
