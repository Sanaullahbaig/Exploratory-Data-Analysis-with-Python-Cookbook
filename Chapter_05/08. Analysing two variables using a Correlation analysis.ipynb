Import relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load dataset

```python
penguins_data = pd.read_csv("penguins_size.csv")
```

Subset for relevant columns

```python
penguins_data = penguins_data[['culmen_length_mm', 'body_mass_g']]
```

Create a Correlation Heatmap in Seaborn

```python
# .corr() method calculates the Pearson correlation coefficient (a measure of linear relationship) 
# between every pair of numerical columns in your penguins_data DataFrame

penguins_corr = penguins_data.corr() # .corr() only works for mathematical values
heatmap = sns.heatmap(penguins_corr, vmin=-1, vmax=1, annot=True)
# annot -> Displays the actual correlation value (the coefficient) on each cell of the heatmap.
```

Set the chart size and chart title. Also set a padding value to define the distance of the title from the top of the heatmap

```python
plt.figure(figsize=(16, 6))

heatmap = sns.heatmap(penguins_data.corr(), vmin=-1, vmax=1, annot=True)

heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':24}, pad=12)
```
